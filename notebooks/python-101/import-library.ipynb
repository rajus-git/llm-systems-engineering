### How to Create a Library / Use Another .py File

#### Project structure
my_project/
    main.py
    utils.py
    calculator.py

##### calculator.py
```
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

PI = 3.14159
```

#### main.py — 3 ways to import it
##### Way 1 — import whole file
````
import calculator
print(calculator.add(3, 4))       # 7
print(calculator.PI)               # 3.14159
````

##### Way 2 — import specific things
````
from calculator import add, PI
print(add(3, 4))                   # 7
````

##### Way 3 — import everything (avoid this ❌)
````
from calculator import *           # pollutes your namespace
````

#### Best practice — use if __name__ == "__main__"
⭐ This is the most important pattern when splitting code across files. Always use it.

````
# calculator.py
def add(a, b):
    return a + b

# This block ONLY runs when you run calculator.py directly
# It does NOT run when another file imports calculator
if __name__ == "__main__":
    print(add(2, 3))   # test code here
````

#### For larger projects — use packages (folders)
````
my_project/
    main.py
    helpers/
        __init__.py     ← this empty file makes it a package
        string_utils.py
        math_utils.py
````

Usage:
````
# main.py
from helpers import string_utils
from helpers.math_utils import add
````
