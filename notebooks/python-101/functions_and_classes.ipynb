## Functions

```python
# Define
def greet():
    print("Hello!")

# Call
greet()   # Hello!
```

```python
# *args — variable number of arguments
def add_all(*args):
    print(args)          # it's a tuple
    return sum(args)

add_all(1, 2, 3)         # 6
add_all(1, 2, 3, 4, 5)   # 15
```

```python
# **kwargs — variable keyword arguments

def print_info(**kwargs):
    print(kwargs)    # it's a dictionary

    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Rayman", city="Hyderabad", age=30)
# name: Rayman
# city: Hyderabad
# age: 30
```

```python
# Order must always be:
# regular → *args → keyword with default → **kwargs
```

```python
# Return multiple values
def min_max(numbers):
    return min(numbers), max(numbers)   # returns a tuple

low, high = min_max([3, 1, 9, 5, 2])
print(low, high)    # 1 9
```

##### Scope — local vs global

```python
x = 10   # global variable

def change():
    x = 99   # local variable — doesn't affect global x
    print(x)  # 99

change()
print(x)     # still 10
```

```python
# To modify a global variable inside a function
x = 10

def change():
    global x
    x = 99

change()
print(x)    # 99
```

⭐ Avoid global as much as possible — it makes code hard to follow. Better to pass values in and return them out.

# Lambda — anonymous one-line function

```python
# Regular function
def square(x):
    return x ** 2

# Lambda equivalent
square = lambda x: x ** 2
print(square(5))   # 25

# Common use — as argument to another function
numbers = [3, 1, 4, 1, 5, 9, 2]
numbers.sort(key=lambda x: x)         # sort normally
print(numbers)

students = [{"name": "Bob", "score": 85},
            {"name": "Alice", "score": 72}]

students.sort(key=lambda s: s["score"])  # sort by score
```
