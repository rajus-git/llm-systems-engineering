## Operators

`````python
# Arithmetic

a = 10
b = 3

a + b    # 13   addition
a - b    # 7    subtraction
a * b    # 30   multiplication
a / b    # 3.333  division (always float)
a // b   # 3    floor division (integer result)
a % b    # 1    modulus (remainder)
a ** b   # 1000 exponent (10 to the power 3)
`````

`````python
# Assignment

x = 10

x += 5    # x = x + 5  → 15
x -= 3    # x = x - 3  → 12
x *= 2    # x = x * 2  → 24
x /= 4    # x = x / 4  → 6.0
x //= 2   # x = x // 2 → 3.0
x **= 2   # x = x ** 2 → 9.0
x %= 4    # x = x % 4  → 1.0
`````

`````python
# Comparison Operators
# always return True or False

a = 10
b = 3

a == b    # False   equal to
a != b    # True    not equal to
a > b     # True    greater than
a < b     # False   less than
a >= b    # True    greater than or equal
a <= b    # False   less than or equal
`````

`````python
# Logical Operators

x = 10

# and — both must be True
x > 5 and x < 20    # True
x > 5 and x < 8     # False

# or — at least one must be True
x > 5 or x > 20     # True
x < 5 or x > 20     # False

# not — flips True/False
not True             # False
not (x > 5)          # False
`````

`````python
# Identity Operators

a = [1, 2, 3]
b = a           # b points to same object
c = [1, 2, 3]   # c is a new object

a is b      # True  — same object in memory
a is c      # False — different objects (even if equal!)
a == c      # True  — same values

a is not c  # True
`````

Tip
⭐ Use == to compare values. Use is to check if two variables point to the same object. Common use: if x is None

`````python
# Membership Operators

fruits = ["apple", "banana", "mango"]

"apple" in fruits       # True
"grape" in fruits       # False
"grape" not in fruits   # True

# Works on strings too
"ello" in "hello"       # True

# Works on dicts (checks keys)
d = {"name": "Raju"}
"name" in d             # True
`````

`````python
# Bitwise Operators

a = 5   # binary: 0101
b = 3   # binary: 0011

a & b   # AND  → 1  (0001)
a | b   # OR   → 7  (0111)
a ^ b   # XOR  → 6  (0110)
~a      # NOT  → -6
a << 1  # left shift  → 10
a >> 1  # right shift → 2
`````

### Operator Precedence
#### Highest → Lowest:

````
**          (exponent)
+x, -x, ~x (unary)
*, /, //, % (multiply/divide)
+, -        (add/subtract)
==, !=, <, >, <=, >= (comparison)
not
and
or
````

# Data Structures



`````python
# List — ordered, changeable collection
fruits = ["apple", "banana", "apple"]  # duplicates allowed
fruits.append("mango")
print(fruits[0])  # "apple"

# accessing 
# By index (starts at 0)
print(fruits[0])   # "apple"
print(fruits[-1])  # "grape"  ← negative index = from the end

# Slicing [start:end]  (end is excluded)
print(fruits[1:3]) # ["banana", "mango"]
print(fruits[:2])  # ["apple", "banana"]
print(fruits[2:])  # ["mango", "grape"]

# Simple loop
for fruit in fruits:
    print(fruit)

# With index — use enumerate()
for i, fruit in enumerate(fruits):
    print(i, fruit)
# 0 apple
# 1 banana ...
`````

`````python
# additional list methods
fruits.append("orange")      # add to end
fruits.insert(1, "kiwi")     # insert at index 1
fruits.remove("banana")      # remove by value
fruits.pop()                 # remove & return last item
fruits.pop(0)                # remove & return at index 0
fruits.sort()                # sort in place
fruits.reverse()             # reverse in place
len(fruits)                  # count of items
"apple" in fruits            # True/False membership check
fruits.count("apple")        # how many times value appears
fruits.index("mango")        # index of a value
`````

`````python
# Tuple — ordered, fixed collection
point = (10, 20)
print(point[1])  # 20

# accessing elements
point = (10, 20, 30, 40)

print(point[0])    # 10
print(point[-1])   # 40
print(point[1:3])  # (20, 30)  ← slicing works same as list
`````

`````python
# iterating
for value in point:
    print(value)

# With index
for i, value in enumerate(point):
    print(i, value)
`````

`````python
# Unpacking — tuples' superpower ⭐
x, y, z, w = point       # unpack all
print(x)  # 10

first, *rest = point      # first + rest
print(first)  # 10
print(rest)   # [20, 30, 40]
`````

`````python
# Tuples have only 2 methods (since they're immutable)
point.count(10)   # how many times 10 appears
point.index(30)   # index of value 30
len(point)        # count of items
`````

`````python
# Set — unique items, no order

visitors = {"Alice", "Bob", "Alice"}  # duplicate removed
print(visitors)  # {"Alice", "Bob"}

# Great for checking membership fast
print("Alice" in visitors)  # True
`````

`````python
# accessing elements

colors = {"red", "blue", "green"}

# ❌ No indexing — sets have no order
# colors[0]  → Error!

# ✅ Only membership check
print("red" in colors)   # True
print("pink" in colors)  # False
`````

`````python
# Simple loop (order is not guaranteed)
for color in colors:
    print(color)
`````

`````python
# Set Methods
colors.add("yellow")          # add one item
colors.update(["pink","white"]) # add multiple
colors.remove("red")          # remove (error if missing)
colors.discard("red")         # remove (no error if missing)
colors.pop()                  # remove a random item

# Set math — very powerful!
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

a | b   # union        → {1,2,3,4,5,6}  (all items)
a & b   # intersection → {3,4}          (common items)
a - b   # difference   → {1,2}          (in a but not b)
a ^ b   # symmetric diff → {1,2,5,6}   (not in both)
`````

`````python
#  Dictionary — key-value pairs
person = {
    "name": "Rayman",
    "city": "Hyderabad",
    "age": 30
}
print(person["name"])  # "Rayman"
person["age"] = 31     # update a value
`````

`````python
# accessing elements
student = {"name": "Rayman", "age": 25, "city": "Hyderabad"}

# By key
print(student["name"])          # "Rayman"
print(student.get("age"))       # 25
print(student.get("phone", "N/A"))  # "N/A" ← default if key missing
# ✅ Prefer .get() over [] to avoid errors
`````

`````python
# Iterating Dictionary
# Keys only
for key in student:
    print(key)

# Values only
for value in student.values():
    print(value)

# Keys AND values together ⭐
for key, value in student.items():
    print(key, "→", value)
# name → Raju
# age  → 25
`````

`````python
# Dictionary Methods
student.keys()           # all keys
student.values()         # all values
student.items()          # all key-value pairs

student["email"] = "r@gmail.com"  # add new key
student["age"] = 26               # update existing key
student.pop("city")               # remove a key

student.update({"age": 27, "phone": "9999"})  # update multiple

"name" in student        # True — check if key exists
len(student)             # number of key-value pairs
`````
