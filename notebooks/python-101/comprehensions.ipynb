Comprehensions are a clean, Pythonic way to create collections in one line.

![alt text](image.png)

```python
# List Comprehension
# Pattern: [expression for item in iterable if condition]

# Basic
squares = [x ** 2 for x in range(6)]
# [0, 1, 4, 9, 16, 25]

# With condition
evens = [x for x in range(10) if x % 2 == 0]
# [0, 2, 4, 6, 8]

# Transform strings
names = ["raju", "alice", "bob"]
upper = [name.upper() for name in names]
# ["RAJU", "ALICE", "BOB"]

# Condition with transformation
scores = [85, 42, 90, 35, 76]
passed = [s for s in scores if s >= 50]
# [85, 90, 76]

```

```python
# if-else inside list comprehension
# Pattern: [value_if_true if condition else value_if_false for item in iterable]

scores = [85, 42, 90, 35, 76]
results = ["pass" if s >= 50 else "fail" for s in scores]
# ["pass", "fail", "pass", "fail", "pass"]
```

```python
# Nested list comprehension

# Flatten a 2D list
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [num for row in matrix for num in row]
# [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Multiplication table
table = [[i * j for j in range(1, 4)] for i in range(1, 4)]
# [[1,2,3], [2,4,6], [3,6,9]]
```

```python
# Dictionary Comprehension

# Pattern: {key: value for item in iterable if condition}

# Basic
squares = {x: x**2 for x in range(5)}
# {0:0, 1:1, 2:4, 3:9, 4:16}

# Flip keys and values
original = {"a": 1, "b": 2, "c": 3}
flipped = {v: k for k, v in original.items()}
# {1:"a", 2:"b", 3:"c"}

# Filter a dictionary
scores = {"Raju": 85, "Alice": 42, "Bob": 90, "Priya": 35}
passed = {name: score for name, score in scores.items() if score >= 50}
# {"Raju": 85, "Bob": 90}
```

```python
# Set Comprehension
# Pattern: {expression for item in iterable}

numbers = [1, 2, 2, 3, 3, 3, 4]
unique_squares = {x**2 for x in numbers}
# {1, 4, 9, 16}  — duplicates removed automatically
```

```python
# Generator Expression
# Like list comprehension but uses () instead of []
# Does NOT build the whole list in memory — computes one at a time

# List comprehension — stores all in memory
squares_list = [x**2 for x in range(1000000)]   # uses lots of memory

# Generator — computes lazily ⭐
squares_gen = (x**2 for x in range(1000000))    # uses almost no memory

# Use when you just need to iterate, not store
total = sum(x**2 for x in range(1000))   # very efficient
print(total)
```
