### Data Types

```python
# check type using type()

# Check type with type()
print(type(42))        # <class 'int'>
print(type(3.14))      # <class 'float'>
print(type("hello"))   # <class 'str'>
print(type(True))      # <class 'bool'>
print(type(None))      # <class 'NoneType'>
```

##### Type Conversion

```python
int("42")        # string → int      → 42
float("3.14")    # string → float    → 3.14
str(100)         # int → string      → "100"
bool(0)          # 0 → bool          → False
bool(1)          # 1 → bool          → True
bool("")         # empty → bool      → False
bool("hi")       # non-empty → bool  → True
```

#### String

```python
name = "Rayman"

# Length
len(name)           # 4

# Access characters
name[0]             # "R"
name[-1]            # "u"

# Slicing
name[1:3]           # "aj"

# Common methods
"hello".upper()     # "HELLO"
"HELLO".lower()     # "hello"
"  hi  ".strip()    # "hi"   ← removes spaces
"hi there".split()  # ["hi", "there"]
"hello".replace("l", "r")  # "herro"

# f-strings — most modern way to format ⭐
city = "Hyderabad"
print(f"My name is {name} and I live in {city}")
# "My name is Rayman and I live in Hyderabad"
```

##### Single, Double, Triple Quotes


```python
# Single and double quotes — NO difference at all
name1 = 'Raju'
name2 = "Raju"   # identical
# The only reason to choose one over the other is to avoid escaping:

msg = 'it\'s a good day'   # need to escape the apostrophe

msg = "it's a good day"    # clean, no escape needed

# Similarly
msg = 'He said "hello"'    # use single to wrap double quotes inside
```

```python
# Multi-line strings
pythonaddress = """
123 Main Street
Hyderabad
Telangana
"""

# 2. Strings with both single and double quotes inside
pythonmsg = """it's a "great" day"""

# 3. Docstrings — documenting functions/classes ⭐
def add(a, b):
    """
    Adds two numbers and returns the result.
    a: first number
    b: second number
    """
    return a + b

# Docstrings are the standard way to document code in Python. Tools like VSCode show them as tooltips.
```
