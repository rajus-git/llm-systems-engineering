# collections module

Python has a module called collections that gives you specialized, more powerful data structures.

```python
from collections import Counter, defaultdict, OrderedDict, deque, namedtuple
```

```python
# Counter — count occurrences automatically


from collections import Counter

words = ["apple", "banana", "apple", "mango", "banana", "apple"]

count = Counter(words)
print(count)
# Counter({'apple': 3, 'banana': 2, 'mango': 1})

# Most common items
print(count.most_common(2))
# [('apple', 3), ('banana', 2)]

# Works on strings too!
letters = Counter("mississippi")
print(letters)
# Counter({'s': 4, 'i': 4, 'p': 2, 'm': 1})

# Use when: you need to count/tally things quickly.
```

```python
# defaultdict — dictionary with a default value

from collections import defaultdict

# Normal dict throws error for missing keys
# d["new_key"] → KeyError ❌

# defaultdict gives a default instead
scores = defaultdict(list)  # default is empty list

scores["Rayman"].append(85)
scores["Rayman"].append(90)
scores["Alice"].append(75)

print(scores)
# {'Rayman': [85, 90], 'Alice': [75]}
# No KeyError even though we never initialized the keys!
```

```python
# With int default (starts at 0)
word_count = defaultdict(int)
for word in ["hi", "hello", "hi", "hi"]:
    word_count[word] += 1

print(dict(word_count))
# {'hi': 3, 'hello': 1}
```

Use when: you want to avoid KeyError and auto-initialize values.

##### deque — double-ended queue (fast insert/remove from both ends)

```python
from collections import deque

d = deque([1, 2, 3])

# Add to either end
d.append(4)        # add to right  → [1,2,3,4]
d.appendleft(0)    # add to left   → [0,1,2,3,4]

# Remove from either end
d.pop()            # remove from right  → [0,1,2,3]
d.popleft()        # remove from left   → [1,2,3]

print(d)  # deque([1, 2, 3])
```

```python
# Rotate — very handy
d = deque([1, 2, 3, 4, 5])
d.rotate(2)   # shift right by 2
print(d)      # deque([4, 5, 1, 2, 3])
```

Use when: you need fast operations on both ends — queues, sliding windows, browser history.

##### namedtuple — tuple with named fields

```python
from collections import namedtuple

# Define structure once
Point = namedtuple("Point", ["x", "y"])
Person = namedtuple("Person", ["name", "age", "city"])

# Create instances
p = Point(10, 20)
person = Person("Rayman", 30, "Hyderabad")

# Access by name OR index
print(p.x)         # 10
print(p[1])        # 20

print(person.name) # "Rayman"
print(person.city) # "Hyderabad"
```

Use when: a plain tuple feels unclear — namedtuple makes code much more readable.

##### OrderedDict — dictionary that remembers insertion order

📝 In Python 3.7+, regular dict also maintains order. So OrderedDict is less commonly needed today — but still useful when order comparison between two dicts matters.

```python
from collections import OrderedDict

od = OrderedDict()
od["first"] = 1
od["second"] = 2
od["third"] = 3

for key, value in od.items():
    print(key, value)
# first 1
# second 2
# third 3
```
