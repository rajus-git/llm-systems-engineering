```python
age = 25

# Basic if/elif/else
if age < 13:
    print("Child")
elif age < 18:
    print("Teenager")
elif age < 60:
    print("Adult")
else:
    print("Senior")
```

```python
# one-liner
age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)   # "Adult"
```

```python
# Checking multiple values cleanly
color = "red"

# Instead of: if color == "red" or color == "blue" or color == "green"
if color in ["red", "blue", "green"]:
    print("Primary color")
```

### Truthy and Falsy 

```python
# Truthy and Falsy 
# These are all treated as False:
if 0:        pass   # False
if "":       pass   # False
if []:       pass   # False
if {}:       pass   # False
if None:     pass   # False

# Everything else is True:
if 1:        print("true")
if "hello":  print("true")
if [1,2]:    print("true")
```

```python
# Practical use
name = input("Enter name: ")
if name:                        # True if not empty
    print(f"Hello {name}")
else:
    print("No name entered")
```

## Loops

##### for

```python
# Loop over a list
fruits = ["apple", "banana", "mango"]
for fruit in fruits:
    print(fruit)

# Loop over a range
for i in range(5):        # 0,1,2,3,4
    print(i)

for i in range(2, 8):     # 2,3,4,5,6,7
    print(i)

for i in range(0, 10, 2): # 0,2,4,6,8  (step=2)
    print(i)

# Loop over string
for char in "Raju":
    print(char)           # R a j u

# Loop with enumerate
for i, fruit in enumerate(fruits):
    print(i, fruit)

# Loop over dictionary
student = {"name": "Raju", "age": 25}
for key, value in student.items():
    print(key, "→", value)
```

##### while

```python
# Runs as long as condition is True
count = 0
while count < 5:
    print(count)
    count += 1          # ⚠️ always update, or infinite loop!

# Classic pattern — keep asking until valid input
while True:
    age = int(input("Enter age: "))
    if age > 0:
        break           # exit loop
    print("Age must be positive, try again")
```

##### loop control

```python
# break — exit loop immediately
for i in range(10):
    if i == 5:
        break
    print(i)            # prints 0 1 2 3 4

# continue — skip current iteration
for i in range(6):
    if i == 3:
        continue
    print(i)            # prints 0 1 2 4 5

# else on a loop — runs only if loop completed WITHOUT break ⭐
for i in range(5):
    if i == 10:
        break
else:
    print("Loop finished normally")  # this prints
```

#### pass — The Do-Nothing Statement
pass is a placeholder that tells Python "there's nothing here yet, but this is intentional."
Python requires a body for certain blocks (functions, loops, classes, conditions). If you leave them empty, it's a syntax error. pass fixes that.
